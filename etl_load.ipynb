```python
import os
import pandas as pd
import sqlite3

# Ensure 'loaded' folder exists
os.makedirs("loaded", exist_ok=True)

# Load transformed data
full_df = pd.read_csv("transformed/transformed_full.csv")
incremental_df = pd.read_csv("transformed/transformed_incremental.csv")

# Save to SQLite
conn_full = sqlite3.connect("loaded/full_data.db")
full_df.to_sql("full_data", conn_full, if_exists="replace", index=False)

conn_incremental = sqlite3.connect("loaded/incremental_data.db")
incremental_df.to_sql("incremental_data", conn_incremental, if_exists="replace", index=False)

# Preview
print("Full Data Preview:")
display(pd.read_sql("SELECT * FROM full_data LIMIT 5", conn_full))

print("Incremental Data Preview:")
display(pd.read_sql("SELECT * FROM incremental_data LIMIT 5", conn_incremental))
```
